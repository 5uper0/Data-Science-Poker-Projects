```python
import pandas as pd
```

```python
df = pd.read_csv('nl25_vs_Player.csv', encoding = 'ISO-8859-1')
```

```python
print(df.head())
print(df.columns)
print(len(df))
```

```python
df['Wins'] = df['Wins'].fillna(0)
df['Losses'] = df['Losses'].fillna(0)
df['Vs..Player'] = df['Vs..Player'].fillna(0)
```

```python
df['HandsVsPlayer'] = df['Wins'] + df['Losses']
df['bbPer100h_VsPlayer'] = (df['Vs..Player'] / .25) / (df['HandsVsPlayer'] / 100)
```

```python
opps_df = df[(df['Wins'] > 0) | (df['Losses'] > 0)]
```

```python
opps_df = opps_df.sort_values('bbPer100h_VsPlayer', ascending=False)
```

```python
l = []
for i, row in opps_df.iterrows():
    if row['Vs..Player'] > 0:
        l.append('x')
    
print(len(l)*.2)
```

```python
opps_df['index_value'] = opps_df.index
```

```python
opps_df['top20VsPlayer'] = 0
for i, row in opps_df.iterrows():
    if i <= 232:
        opps_df['top20VsPlayer'].loc[i] = 1        
```
